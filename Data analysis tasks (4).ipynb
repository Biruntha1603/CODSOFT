```python
print("Task 1 : Student Marks Dashboard")
students = [
    ["Alice", 85, 90, 95],
    ["Bob", 75, 80, 82],
    ["Carol", 92, 88, 91],
    ["David", 60, 65, 70]
]

print("\n1. Total Marks for Each Student:")
for student in students:
    name = student[0]
    total_marks = student[1]+student [2]+student [3]
    print(f"{name}: {total_marks}")

highest_avg = 0
student_name =""
for student in students:
    name = student[0]
    total_marks = student[1] + student[2] + student[3]
    avg= total_marks / 3
    if avg > highest_avg:
        highest_avg = avg
        student_name = name
print(f"\n2. Student with the Highest Average: {student_name}")

print("\n3. Students who scored more than 250 in total:")
for student in students:
    name = student[0]
    total_marks =student [1] +student [2]+student [3]
    if total_marks > 250:
        print(name)

print("\n4. Adding a new student and reprinting all records:")
new= ["Eve", 70, 75, 80]
students.append(new)
for student in students:
    print(student)
```

```python
print("Task 2:Hobby Tracker")
hobby_data = [
    ["Alice", "reading", "cycling"],
    ["Bob", "drawing", "football"],
    ["Carol", "reading", "swimming"],
    ["David", "cycling", "gaming"]
]

unique_hobbies = set()
for hobby in hobby_data:
    unique_hobbies.add(hobby[1])
    unique_hobbies.add(hobby[2])
print(f"\n1. All Unique Hobbies: {unique_hobbies}")

reading_count = 0
for hobby in hobby_data:
    if 'reading' in hobby:
        reading_count += 1
print(f"\n2. Students with 'reading' as a hobby: {reading_count}")

print("\n3. Students who share at least one hobby:")
shared_hobby_pairs = set()
for i in range(len(hobby_data)):
    for j in range(i + 1, len(hobby_data)):
        student1_name = hobby_data[i][0]
        student1_hobbies = {hobby_data[i][1], hobby_data[i][2]}
        
        student2_name = hobby_data[j][0]
        student2_hobbies = {hobby_data[j][1], hobby_data[j][2]}
        
        if student1_hobbies.intersection(student2_hobbies):
            pair = tuple(sorted((student1_name, student2_name)))
            shared_hobby_pairs.add(pair)
for pair in shared_hobby_pairs:
    print(f"{pair[0]} and {pair[1]} share at least one hobby.")

print("\n4.Dictionary")
dict={hobby[0]:hobby[1:] for hobby in hobby_data}
print(dict)
```

```python
print("Task 3: Product Sales Matrix")

sales = [
    ["Rice", 120, 130, 125],
    ["Wheat", 100, 90, 95],
    ["Oil", 80, 85, 90],
    ["Sugar", 110, 115, 105]
]
print("\n1.Total sales for each product:")
for item in sales:
    product = item[0]
    sale=item[1]+item[2]+item[3]
    print(f"{product}:{sale}")

print("\n2.Product with highest sales in march:")
highest_sales=0
product_name=""
for item in sales:
    product =item[0]
    mar_sales=item[3]
    if mar_sales>highest_sales:
        highest_sales =mar_sales 
        product_name=product 
        print(product)
print("\n3.Monthwise total sales:")
jan_sales=sum(item[1] for product in sales)
feb_sales=sum(item[2] for product in sales)
mar_sales=sum(item[3] for product in sales)
print(f"January sales:{jan_sales}")
print(f"February sales:{feb_sales}")
print(f"March sales:{mar_sales}")

sales_dict = {product[0]: product[1:] for product in sales}
print("\n4.Dictionary:")
print(sales_dict)


```

```python
print("Task 4:Library book manager")
books = [
    ["Wings of Fire", "A.P.J. Abdul Kalam", "Biography"],
    ["1984", "George Orwell", "Fiction"],
    ["The Alchemist", "Paulo Coelho", "Fiction"],
    ["Zero to One", "Peter Thiel", "Business"]
]

print("\n1.Books under the 'Fiction' genre:")
for book in books:
    if book[2] == "Fiction":
        print(book[0])

print("\n2.Unique authors:")
unique_authors = sorted(list(set(book[1] for book in books)))
print(unique_authors)

print("\n3.Book counts per genre:")
genre_counts = {}
for book in books:
    genre = book[2]
    genre_counts[genre] = genre_counts.get(genre, 0) + 1
for genre, count in genre_counts.items():
    print(f"{genre}: {count}")

print("\n4.Converted list:")
dicts=[]
for book in books:
    dicts.append({
        'title': book[0],
        'author': book[1],
        'genre': book[2]
    })
print(dicts)
```

```python
print("Task 5: Employee Skills Directory")

employees = [["Arun", "Python", "SQL", "Excel"],
             ["Divya", "Tableau", "Python", "Power BI"], 
             ["Rahul", "Excel", "Communication", "SQL"], 
             ["Meena", "Python", "ML", "DL"]
]
print("\n1. Employees skilled in 'Python':")
python_skill = []
for skills in employees:
    if "Python" in skills[1:] :
       python_skill.append(skills[0])
print(python_skill)

print("\n2.Count frequency of each skills:")
skill_frequency = {}
for employee in employees:
    for skill in employee[1:]:
        skill_frequency[skill] = skill_frequency.get(skill, 0) + 1
for skill, count in skill_frequency.items():
    print(f" {skill}: {count}")

print("\n3.Skill wise index:")
skill_index = {}
for employee in employees:
    employee_name = employee[0]
    for skill in employee[1:]:
        if skill not in skill_index:
            skill_index[skill] = []
            skill_index[skill].append(employee_name)
for skill, employee_list in skill_index.items():
    print(f"'{skill}': {employee_list}")

print("\n4.Checking for employees with 'Python','SQL','Excel'")
required_skills = {"Python", "SQL", "Excel"}
employee_with_all_skills = False
for employee in employees:
    employee_name = employee[0]
    employee_skills = set(employee[1:])
    if required_skills.issubset(employee_skills):
        print(f"{employee_name} has all three skills: Python, SQL, Excel")
        employee_with_all_skills = True
if not employee_with_all_skills:
    print("No employee found with all three skills: Python, SQL, Excel.")
```

```python
print("Task 7:Installment Tracker")
payments = [
    ["Arun", 600000, 300000, 300000],
    ["Divya", 750000, 750000, 0],
    ["Meena", 500000, 100000, 400000],
    ["Kumar", 900000, 600000, 300000]
]
print("\n1.Names of customers with dues")
dues=[]
for payment in payments:
    if payment[3]>0:
        dues.append(payment[0])
print(dues)

print("\n2.Each customer paid percentage")
for payment in payments:
    percentage = payment [2]/payment [1]*100
    print(f"{payment[0]} paid percentage is {percentage}")


print("\n3.Fully paid customers")
paid=[]
for payment in payments:
    if payment [1]==payment [2]:
        paid.append(payment [0])
print(paid)

print("\n4.Update a payment and recalculate dues")
customer_to_update = "Meena"
new_paid_amount = 250000 
for i,payment in enumerate(payments):
    if payment[0] == customer_to_update:
          total_amount = payment[1]
          payments[i][2] = new_paid_amount  
          payments[i][3] = total_amount - new_paid_amount 
          print(f"Updated payment for {customer_to_update}:{payments[i]}")

        
```

```python
print("Task 8:Plot Location & Facing Info")
plot_info = [
    ["P101", "North Layout", "East"],
    ["P102", "North Layout", "West"],
    ["P103", "South Layout", "North"],
    ["P104", "East Layout", "East"]
]

print("\n1.List of plots in north layout:")
plots=[]
for plot in plot_info:
    if plot[1]== "North Layout":
        plots.append(plot[0])
print(plots)  

print("\n2.Plots facing east:")
plots =[]
for plot in plot_info:
    if plot[2]=="East":
        plots.append(plot[0])
print(plot)

print("\n3. Search for a plot by ID:")
plot_id = "P103"
plots = None
for plot in plot_info:
    if plot[0] == plot_id:
        plots = plot
        break
if plots:
    print(f"Details for Plot ID '{plot_id}':")
    print(f"Location: {plots[1]}")
    print(f"Facing: {plots[2]}")
else:
    print(f"Plot ID '{plot_id}' not found.")

print("\n4. Group plots by layout location:")
plots_by_layout = {}
for plot in plot_info:
    plot_id = plot[0]
    location = plot[1]
    if location not in plots_by_layout:
        plots_by_layout[location] = []
    plots_by_layout[location].append(plot_id)

for location, plot_ids in plots_by_layout.items():
    print(f"{location}: {plot_ids}")
```
